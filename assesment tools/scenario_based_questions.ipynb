```python
import numpy as np

house_data = np.array([
    [3,1400,7500000],
    [5,2200,12500000],
    [4,1800,9800000],
    [6,3000,18500000],
    [2,1100,5200000]
])

# Get minimum bedrooms from user input
min_bedrooms = int(input("Enter minimum bedrooms: "))

# Filter houses based on user input
filtered_houses = house_data[house_data[:,0] > min_bedrooms]

# Check if any houses match the criteria
if len(filtered_houses) > 0:
    average_price = np.mean(filtered_houses[:,2])
    print(f"Average Sale Price for houses with more than {min_bedrooms} bedrooms = ₹{average_price:,.2f}")
else:
    print(f"No houses found with more than {min_bedrooms} bedrooms.")
```

```python
from sklearn.linear_model import LogisticRegression

X = [
    [1,3,120],
    [6,15,300],
    [0,1,80],
    [4,10,250],
    [2,5,160]
]

y = [0,1,0,1,0]

model = LogisticRegression()
model.fit(X, y)

# New email
new_email = [[5,12,280]]

prediction = model.predict(new_email)

if prediction[0] == 1:
    print("Spam")
else:
    print("Not Spam")
```

```python
from sklearn.linear_model import LogisticRegression

X = [
    [1,3,120],
    [6,15,300],
    [0,1,80],
    [4,10,250],
    [2,5,160]
]

y = [0,1,0,1,0]

model = LogisticRegression()
model.fit(X,y)

new_email = [[3,8,200]]

prediction = model.predict(new_email)

print("Prediction =", prediction[0])

if prediction[0] == 1:
    print("Spam")
else:
    print("Not Spam")
```

```python
from sklearn.linear_model import LinearRegression

# Brand Encoding:
# Toyota=1, Hyundai=2, Honda=3, Maruti=4, Tata=5

X = [
    [1,2,25000,1200],
    [2,5,60000,1500],
    [3,3,40000,1300],
    [4,7,90000,1800],
    [5,1,15000,1000]
]

y = [650000,420000,520000,300000,720000]

model = LinearRegression()
model.fit(X, y)

brand = 3      # Honda
age = 4
kms = 50000
engine = 1400

prediction = model.predict([[brand, age, kms, engine]])

print("Predicted Resale Price = ₹", prediction[0])
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text

X = [
    [45000,720,0],
    [30000,650,1],
    [60000,780,0],
    [28000,620,2],
    [52000,710,1]
]

y = [1,0,1,0,1]

model = DecisionTreeClassifier(criterion="gini", random_state=0)
model.fit(X,y)

new_applicant = [[50000,730,1]]

prediction = model.predict(new_applicant)

if prediction[0] == 1:
    print("Loan Approved")
else:
    print("Loan Rejected")

print(export_text(model,
                  feature_names=["Income","Credit_Score","Existing_Loans"]))
```
